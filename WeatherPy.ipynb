```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint 

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding geocodes and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Perform a weather check on each city using a series of successive API calls.
# Include a print log of each city as it'sbeing processed (with the city number and city name).


# citi_test = [cities[0], cities[1], cities[2], cities[3]]

# api.openweathermap.org/data/2.5/weather?q={city name}

base_url = "http://api.openweathermap.org/data/2.5/weather?"
City_Name = []
units = 'imperial'
Cloudiness = []
Country = []
Date =  []
Humidity = []
Lat = []
Lng = []         
Max_Temp = []      
Wind_Speed = []   

# for city in citi_test:
for city in cities: 
#     print(city)
    url = f"{base_url}appid={api_key}&units={units}&q={city}"
    city_info = requests.get(url).json()

    try:
        print(f"Storing weather data for {city_info['name']} index[{cities.index(city)}].")
        
        City_Name.append(city_info['name'])      
        Country.append(city_info['sys']['country'])    
        Date.append(city_info['dt'])
        Max_Temp.append(city_info['main']['temp_max'])
        Humidity.append(city_info['main']['humidity'])
        Cloudiness.append(city_info['clouds']['all'])
        Wind_Speed.append(city_info['wind']['speed']) 
        Lat.append(city_info['coord']['lat'])
        Lng.append(city_info['coord']['lon'])
    except (KeyError, IndexError):
        print(f"Missing field/result for {city} index[{cities.index(city)}] ... skipping.")
        print("------------")    


# print(Cloudiness[0:9])
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_data = {
'City': City_Name,
'Country': Country, 
'Date': Date,
'Max Temp': Max_Temp, 
'Humidity': Humidity,
'Cloudiness': Cloudiness,
'Wind Speed': Wind_Speed,
'Lat': Lat,
'Lng': Lng      
}

city_df = pd.DataFrame.from_dict(city_data)

city_df.to_csv('city_data.csv', index = None, header=True)
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
city_date = time.strftime('%m/%d/%y', time.localtime(city_df['Date'][0]))
```

```python
plt.scatter(city_df['Lat'], city_df['Max Temp'], marker='o', facecolor='cornflowerblue', edgecolor= 'black', alpha=.75)
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')

plt.title(f'Max Temperature ({city_date}) vs. City Latitude')
plt.savefig('Temp vs Latitude')
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(city_df['Lat'], city_df['Humidity'], marker='o', facecolor='cornflowerblue', edgecolor= 'black', alpha=.75)
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')

plt.title(f'Humidity ({city_date}) vs. City Latitude')
plt.savefig('Humidity vs Latitude')
```
